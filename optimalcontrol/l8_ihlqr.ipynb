# Infinite Horizon LQR

For time-invariant LQR, the K and P matrices converge to constants.

This is useful for many stabilization problems (like quadrotor hover) as we can use a single K.

$$
\begin{aligned}
& K_k=\left(R+B^T P_{k+1} B\right)^{-1} B^T P_{k+1} A \\
& P_k=Q+A^T P_{k+1}\left(A-B K_k\right)
\end{aligned}
$$

In the infinite horizon case 

$$
\implies P_k \rightarrow P_{k+1} \rightarrow P_{inf}
$$

With Julia/Matlab/Python the `dare` (Discrete Algebraic Riccati Equation) function finds the $K_{inf}$ for you. eg `Kinf = dare(A,B,Q,R)`.
